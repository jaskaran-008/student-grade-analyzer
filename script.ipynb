```python
student_list = []

try :
    with open('students.csv','r') as f:
        content = f.readlines()
        lines = content[1:]

        for line in lines:
            parts = line.split(',')

            name = parts[0]
            m = int(parts[1])
            s = int(parts[2])
            e = int(parts[3])
            st = int(parts[4])

            student_dict = {
                'Name':name,
                'Maths':m,
                'Science':s,
                'English':e,
                'Social Studies':st
            }

            student_list.append(student_dict)

except FileNotFoundError:
    print('File not found')

student_list
```

```python
def grading(num):
    if(num >= 90):
        return 'A'
    elif(num >= 85 and num < 90):
        return 'B'
    elif(num >= 80 and num < 85):
        return 'C'
    else:
        return 'F'
    

for record in student_list:
    try :
        total_marks = int(record['Maths']) + int(record['Science']) + int(record['English']) + int(record['Social Studies'])

        average = int(total_marks/4)
        record['Average'] = average

        percentage = float((total_marks/400)*100)
        record['Percentage'] = percentage

        grade =  grading(percentage)
        record['Grade'] = grade
        

    except ValueError:
        print('Marks of Student is Missing')

for s in student_list:
    print(s['Name'],' Percentage : ',s['Percentage'], ' Grade : ',s['Grade'])
```

```python
ranking = sorted(student_list, key = lambda x : x['Percentage'], reverse=True)
ranking

print('Top 3 toppers are : ')
for top in ranking[:3]:
    print(top['Name'],', Percentage',top['Percentage'])
    
topper_name = ranking[0]['Name']
```

```python
list_sub = ['Maths','Science','English','Social Studies']

subject_average = {subject : sum( sub[subject] for sub in student_list )/len(student_list) for subject in list_sub}
print(subject_average)

highest_scoring = max(subject_average.items())
lowest_scoring = min(subject_average.items())

print('Highest Scoring Subject in Class : ', highest_scoring)
print('Lowest Scoring Subject : ',lowest_scoring)
```

```python
passed_students = list(filter(lambda x : x.get('Grade') != 'F', ranking))

for p in ranking:
    if(p['Grade'] == 'F'):
        p['Result'] = 'Fail'
    else :
        p['Result'] = 'Pass'

students_count = len(student_list)
passed_count = len(passed_students)
failed_count = students_count - passed_count

passsed_percent = (passed_count/ students_count)*100

print('Number of Students Passed : ',passed_count)
print('Number of Student Failed : ',failed_count)
print('Passed Percentage : ', passsed_percent,'%')
```

```python
from datetime import datetime
import os

folder_name = 'Final_Report'
os.makedirs(folder_name, exist_ok=True)
file_path = os.path.join(folder_name,'student_report.txt')

with open(file_path,'w') as f:
    f.write('Students Grade Report - ')
    f.write(datetime.now().strftime('%Y-%m-%d'))
    f.write('\n___________________________________\n')
    f.write('\nName | Percentage | Grade | Result \n')

    for final in ranking:

        line = f"{final['Name']} | {final['Percentage']}\t\t| {final['Grade']} | {final['Result']}\n"
        f.write(line)

    f.write('\n___________________________________\n')
    f.write('\n\nClass Pass Rate : ')
    p_p = f"{passsed_percent}"
    f.write(p_p)

    f.write('\n\nTopper Name : ')
    f.write(topper_name)

    f.write('\n\nHighest Scoring Subject of Class with Average :')
    h_s = f"{highest_scoring}"
    f.write(h_s)

    f.write('\n\nLowest Scoring Subject of Class with Average :')
    l_s = f"{lowest_scoring}"
    f.write(l_s)
    
    

    




```
